# Read data

```python
# read data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# expectations
data_expectation = pd.read_csv('Expectations about contributing to Open Source Survey Student Analysis Report(1).csv')
data_expectation
```

```python
# impressions
data_impression = pd.read_csv('Impressions about contributing to Open Source - after Survey Student Analysis Report(1).csv')
data_impression
```

# Preprocessing

```python
# remove numbers from each column name like this: 1947787: abcdddd -> abcdddd
cleaned_column_names = []
for column in data_expectation.columns:
    cleaned_column_names.append(column.split(':')[1])
data_expectation.columns = cleaned_column_names
data_expectation
```

```python
cleaned_column_names = []
for column in data_impression.columns:
    cleaned_column_names.append(column.split(':')[1])
data_impression.columns = cleaned_column_names
data_impression
```

# Visualisation

```python
# show data distribution of the 2nd colomn
col2_expectation = data_expectation.iloc[:, 1]
col2_impression = data_impression.iloc[:, 1]

# count the number of each option
col2_expectation.value_counts()
col2_impression.value_counts()

# plot the data distribution into one figure parrallelly
categories = col2_expectation.value_counts().index
expectation = col2_expectation.value_counts()
impression = col2_impression.value_counts()
impression = impression.reindex(categories, fill_value=0)

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plotting bars
rects1 = ax.bar(x - width/2, expectation, width, label='Expectation')
rects2 = ax.bar(x + width/2, impression, width, label='Impression')

# Adding labels, title, and legend
ax.set_xlabel('Responses')
ax.set_ylabel('Count')
ax.set_title('Comparison of Comfort Asking for Help from Open Source Community')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=0)
ax.legend()

# Adding value labels on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```
